```python
years = []
months = []

# Change date range as required
for year in range(2009, 2024):
    for month in range(1, 13):
        years.append(year)
        months.append(month)

df_ym = pd.DataFrame({'Year': years, 'Month': months})

def get_start_end_dates(row):
    start_date = pd.Timestamp(f"{row['Year']}-{row['Month']}-01")
    end_date = start_date + pd.offsets.MonthEnd(0)
    return start_date, end_date

df_ym[['Start_Date', 'End_Date']] = df_ym.apply(get_start_end_dates, axis=1, result_type='expand')

df_dates = df_ym.copy()
df_dates
```

```python
start_date_list = df_dates['Start_Date'].tolist()
# start_date_list

end_date_list = df_dates['End_Date'].tolist()
# end_date_list
```

```python
from pytrends.request import TrendReq
import pandas as pd
import time
pytrend = TrendReq()

df_dates = pd.DataFrame({"Start_Date": start_date_list, "End_Date": end_date_list})

interest_over_time_df = pd.DataFrame()

pytrends = TrendReq(hl='en-US', tz=360)


for k in range(len(df_dates)):
    start_date = str(df_dates[["Start_Date", "End_Date"]].iloc[k][0])[:10]
    end_date = str(df_dates[["Start_Date", "End_Date"]].iloc[k][1])[:10]
    print(start_date, end_date)

    success = False
    while not success:
        try:
            if(start_date == '2023-12-01'):
              break
            pytrend.build_payload(kw_list=[symbol], timeframe=f'{start_date} {end_date}')
            interest_over_time_df = interest_over_time_df.append(pytrend.interest_over_time())
            success = True
        except Exception as e:
            print(f"Future Date Accessed: {e}")
            print("Retrying...")
            time.sleep(5)

interest_over_time_df
```

```python
interest_over_time_df.rename(columns={symbol : 'Search_Trend'}, inplace = True)
interest_over_time_df = interest_over_time_df[['Search_Trend']]
interest_over_time_df

# interest_over_time_df.to_csv('Google_Trends.csv')
```
